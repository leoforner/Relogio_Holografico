```python
from PIL import Image
import requests
from PIL import ImageEnhance
from io import BytesIO
import numpy as np

def redimensionar_e_alterar_imagem(url, largura, altura, fator_contraste):
    # Baixa a imagem da URL
    response = requests.get(url)
    response.raise_for_status()  # Verifica se o download foi bem-sucedido

    # Abre a imagem com Pillow
    img = Image.open(BytesIO(response.content))

    # Redimensiona a imagem
    img = img.resize((largura, altura), Image.LANCZOS)  # LANCZOS para melhor qualidade

        # Aumento de contraste usando ImageEnhance
    enhancer = ImageEnhance.Contrast(img)
    img = enhancer.enhance(fator_contraste)

    # Converte a imagem para o modo RGB (caso não esteja)
    img = img.convert('RGB')

    # Obtém os pixels da imagem
    pixels = img.load()

    # Converte para preto e branco
    pb = img.convert('L')
    pixels_pb = pb.load()

    # Percorre todos os pixels
    for x in range(img.width):
        for y in range(img.height):
            r, g, b = pixels[x, y]

            # Altera valores menores que 20 para 255
            if  pixels_pb[x,y] > 200:
                pixels[x, y] = (0, 0, 0)


    array_pixels = np.array(img)

    for linha in array_pixels:
      for pixel in linha:
          print("{" + ",".join(map(str, pixel)) + "}", end=",")

    return img

# Exemplo de uso
url_imagem = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcS2kYdEwVOHZ19GWO-S6m0W78ZN-8vNuwwx7Q&s"  # Substitua pela URL da sua imagem
largura_desejada = 36
altura_desejada = 36

imagem_alterada = redimensionar_e_alterar_imagem(url_imagem, largura_desejada, altura_desejada, 2 )

# Salva a imagem alterada (opcional),
imagem_alterada.save("imagem_alterada.jpg")

# Exibe a imagem alterada (opcional)
imagem_alterada.show()



```
